# Pandas - Change Column Names

```python
import pandas as pd
```

```python
df = pd.read_csv('iris.data', names=['A','B','C','D','Label'])
df
```

axis=1 tells Pandas it's column names.  
inplace=True tells Pandas to save the changes to our DataFrame.

```python
df.rename({'A':'aa', 'B':'bb'}, axis=1, inplace=True)
df
```
