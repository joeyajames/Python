# Pandas - Delete Columns from a DataFrame

```python
import pandas as pd
```

```python
df = pd.read_csv('iris.data', names=['A','B','C','D','Label'])
df
```

### 1) to drop a single column
, and saving it permanently df.drop('col_name', axis=1,inplace =True)  
To save changes you must either set df = df.drop(), or add inplace=True.

```python
df.drop('B', axis=1)
```

```python
df
```

### 2) to drop multiple axes by name
df.drop(['col_name1', 'col_name2'], axis=1, inplace=True)

```python
df.drop(['A','Label'], axis=1, inplace=True)
df
```

### 3) to drop multiple axes by numerical column index
df.drop(df.columns[[idx1, idx2]], axis=1, inplace=True)

```python
df.drop(df.columns[[0, 2]], axis=1, inplace=True)
df
```
