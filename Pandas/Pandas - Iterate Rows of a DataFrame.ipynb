# Pandas - Iterate Rows of a DataFrame 

```python
import pandas as pd
```

```python
df = pd.read_csv('iris.data', header=None)
df
```

### Get a List of one Column
Use a list comprehension.

```python
col2 = [x for x in df[2]]
col2
```

### Iterate Rows with Index and each Row as a List
**DO NOT** try to change data in the df this way, but it is convenient for iterating.  
Itertuples is supposed to be much faster than Iterrows for large datasets.

```python
for i, row in df.iterrows():
    print(i, row[0], row[1], row[2], row[3], row[4])
```

```python
for row in df.itertuples(name=None): 
    print(row[0], row[1], row[2], row[3], row[4], row[5])
```
